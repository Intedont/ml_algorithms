```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([5.3,6.3,4.8,3.8, 3.3 ,3, 2, 1.6])

plt.plot(x, y, 'ro')
plt.show()

def line(a, b, x):
  return(a*x + b)

def grad_desc_line(n_epoch, lr, a, b):
  for i in range(n_epoch):
    grad_a = np.sum(-2 * x * (y - a*x - b)) / x.shape
    grad_b = np.sum(-2 * (y - a*x - b)) / x.shape
  
    a = a - lr * grad_a
    b = b - lr * grad_b

    if(i % 100 == 0):
      plt.plot(x, y, 'ro')
      plt.plot(x, line(a,b,x))
      plt.show()
      
  return(a, b)
```

```python
grad_desc_line(400, 0.02, 4, 2)
```
